# First Prediction

```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, accuracy_score, f1_score, precision_score, recall_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
```

```python
df = pd.read_csv("processed_data/numerical.csv", index_col='EmployeeID')
df
```

```python
X = df[df.columns.difference(['Attrition_num'])]
y = df['Attrition_num']

y.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
y_test.value_counts()
```

```python
y_train.value_counts()
```

```python
# Basic clasificator training
```

```python
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X_train, y_train)
```

```python
X_test
```

```python
y_prediction = neigh.predict(X_test)
```

```python
y_prediction
```

```python
plt.rcParams['figure.figsize'] = 10, 5

print(f1_score(y_test, y_prediction, average='micro'))
print(precision_score(y_test, y_prediction, average="macro"))
print(recall_score(y_test, y_prediction, average="macro"))

print(classification_report(y_test,y_prediction))
ConfusionMatrixDisplay.from_predictions(y_test, y_prediction)
plt.show()
```
